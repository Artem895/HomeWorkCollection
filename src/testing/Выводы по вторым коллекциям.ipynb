```python
import pandas as pd
import matplotlib.pyplot as plt
```

<h1>Добавление ArrayList LinkedLIst</h1>

все время в микро секундах , использовался метод System.nanotime().Здесь время это время добавдения какого-то фиксированного количества элементов , индексы это количество попыток 

```python
aadd=pd.read_csv("a.csv")
```

```python
aadd
```

Уберем явный выброс в данных , для большей наглядности в графиках 

```python
addnorm=aadd.iloc[1:]
```

```python
y=addnorm.ArrayList
y1=addnorm.LinkedList
x=addnorm.index
fig, ax = plt.subplots()
ax.plot(x, y,
        color = 'r',
        linewidth = 3)
ax.plot(x, y1,
        color = 'g',
        linewidth = 3)
ax.grid(which='major', 
        color = 'k', 
        linewidth = 2)

ax.minorticks_on()
ax.grid(which='minor')
fig.set_figwidth(20)
fig.set_figheight(12)
ax.set_xlabel('номер  ')
ax.set_ylabel('T, микросек ')
ax.set_title('Добавление',fontsize = 20)
##ax.xaxis.set_major_locator(LinearLocator(30))
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
     label.set_fontsize(13)              
plt.show()
```

 <span style="color:red">ArrayList </span><br>
 <span style="color:green">LinkedList </span>

Как видно из графика ArrayList в общем быстрее на добавление в конец , чем LinkedList.Эта разница ни такая уж и большая , следовательно в большей степени безразницы что использовать.Но если нам важна каждая микросекунда , то лучше использовать ArrayList

HashSet и LinkedHashSet почти одинаковы по времени добавления элементов , а вот TreeSet получился самым медленным из всех Set.Следовательно елси нам важно время добавления то лучше использовать HashSet или LinkedHashSet.
