<h1>Добавление элементов</h1>

Датасет data1 представляет собой набор из времени добавления 10 элемнтов типа String (первые 10 элемнтов) в милисекундах и 
времени добавления 200 элемнтов типа String (последние  10 элемнтов) в милисекундах

```python
dataabotadd=pd.read_csv("data1.csv")
```

```python
addtenstr=dataabotadd.head(10)
```

```python
dataabotadd.tail(10)
```

```python
addtwohundredsstr=dataabotadd.tail(10)
```

```python
addtwohundredsstr
```

```python
y=addtenstr.MyLinkedList
x=addtenstr.index
y1=addtenstr.LinkedList
fig, ax = plt.subplots()
ax.plot(x, y,
        color = 'r',
        linewidth = 3)
ax.plot(x, y1,
        color = 'g',
        linewidth = 3)
ax.grid(which='major', 
        color = 'k', 
        linewidth = 2)

ax.minorticks_on()
ax.grid(which='minor')
fig.set_figwidth(20)
fig.set_figheight(12)
ax.set_xlabel('номер эксперемента')
ax.set_ylabel('t,милисекунды')
ax.set_title('Добавление ,10 элементов типа String',fontsize = 20)
##ax.xaxis.set_major_locator(LinearLocator(30))
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
     label.set_fontsize(13)
ax.text(4,1.4, 'LinkedList').set_fontsize(23)   
ax.text(1.5, 15, 'MyLinkedList').set_fontsize(23)                 
plt.show()
##plt.savefig('Laba13g.png')
```
